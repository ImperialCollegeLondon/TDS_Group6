```python
# Create the heatmap
fig, ax = plt.subplots(figsize=(10,10))
cmap = sns.diverging_palette(220, 10, as_cmap=True)
sns.heatmap(retic, cmap=cmap, annot=True, fmt='.4f', ax=ax, cbar_kws={'label': 'Value'})

# Mask the values below 0.05
mask = np.zeros_like(retic)
mask[retic < 0.05] = True
sns.heatmap(retic, cmap=cmap, annot=True, fmt='.4f', ax=ax,
            cbar_kws={'label': 'Value'}, mask=mask, vmin=0, vmax=1, alpha=0.3, annot_kws={'color': 'white'})

# Set the title and axis labels
ax.set_title('Cluster Heatmap')
ax.set_xlabel('Clusters')
ax.set_ylabel('Clusters')

# Show the plot
plt.show()
```
